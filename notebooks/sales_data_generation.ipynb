```python
import pandas as pd
import random

products = [
    "Laptop",
    "Monitor",
    "Keyboard",
    "Mouse",
    "Headphones"
]

categories = {
    "Laptop":"Electronics",
    "Monitor":"Electronics",
    "Keyboard":"Electronics",
    "Mouse":"Electronics",
    "Headphones":"Electronics"
}

rows = []

for i in range(1000):

    product = random.choice(products)

    rows.append({
        "order_id": i + 1,
        "product": product,
        "category": categories[product],
        "amount": random.randint(500,50000)
    })

df = pd.DataFrame(rows)

df.head()
```

```python
df.to_csv("sales.csv", index=False)
```
